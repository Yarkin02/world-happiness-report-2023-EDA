### Veriyi Tanıma ve Analiz

* **Country Name :** Ülke Adı

* **Ladder Score :** Mutluluk Skoru, yaşam değerlendirmesi sorusuna verilen ulusal ortalama yanıt.

* **Logged GDP Per Capita :** Kişi başına düşen Gayri Safi Yurtiçi Hasıla(GSYH)

* **Social Support :** Sosyal Destek, bireye sosyal ağ üyeleri tarafından sağlanan yardım veya destek.

* **Healty Life Expectancy :** Sağlıklı yaşam beklentisi, bir kişinin sağlıklı bir durumda yaşamaya devam etmesi beklenen yıl sayısı.

* **Freedom to make life choices :** Yaşam seçimleri yapma özgürlüğü, "Hayatınızda ne yapacağınızı seçme 
özgürlüğünüzden memnun musunuz, yoksa memnun değil misiniz?" sorusuna verilen ikili yanıtların 
ulusal ortalaması. 

* **Generosity :** "Son bir ay içinde bir hayır kurumuna bağışta bulundunuz mu?" sorusuna verilen ortalama yanıtların kişi başına düşen GSYH üzerindeki regresyonun artıklarını ifade eder.

* **Perceptions of corruption :** Yolsuzluk algısı ifadesi, "Hükümetin genelinde yolsuzluk yaygın mı, değil mi?" ve "İşletmeler arasında yolsuzluk yaygın mı, değil mi?" iki soruya verilen anket yanıtlarının ulusal ortalaması.

* **Ladder Score in Dystopia :** Distopya, ideal bir toplumun tam tersi, genellikle korkunç ve baskıcı bir geleceği tasvir eden bir kurgusal kavramdır. Distopya mutluluk skoru 1.778 olarak belirtilmiş mutluluk skoru bu değere yaklaşmış
veya bu değerden küçük olan ülkeler distopya olarak kabul edilir.

```python
# 2023'De En Mutlu ve En Mutsuz Ülkeler

# Literatürde yapılan değerlendirmelere göre mutluluk endeksi 7'nin üzerinde olan ülkede yaşayan insanlar mutlu,
# mutluluk endeksi 4'ün altında olan ülkede yaşayan insanlar mutsuz kabul ediliyor.

happiest = df[(df.loc[:, "Ladder score"] > 7)]
unhappiest = df[(df.loc[:, "Ladder score"] < 4)]

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
sns.barplot(x = "Ladder score", y = "Country name", data = happiest, palette = "flare")
plt.title("En Mutlu Ülkeler")
plt.xlabel("Mutluluk Skoru")
plt.ylabel("Ülke Adı")

plt.subplot(2, 3, 2)
sns.barplot(x = "Ladder score", y = "Country name", data = unhappiest, palette = "crest")
plt.title("En Mutsuz Ülkeler")
plt.xlabel("Mutluluk Skoru")
plt.ylabel("Ülke Adı")

plt.tight_layout()
```

```python
from scipy.stats import shapiro

print(shapiro(df["Ladder score"]))
print(shapiro(df["Perceptions of corruption"]))
print(shapiro(df["Social support"]))
print(shapiro(df["Logged GDP per capita"]))
print(shapiro(df["Freedom to make life choices"]))
print(shapiro(df["Healthy life expectancy"]))

# Sadece mutluluk skoru p-value değeri 0.05'den büyük olduğu için normal dağılmıştır diyebiliriz.
# Diğer değişkenler için p-value değerleri 0.05'den küçüktür. Veriler normal dağılımamıştır.
```

```python
#Normallik şartı sağlanmadığı için non parametrik korelasyon testi olan spearmanr kullanılmıştır.

# H0 : p = 0
# H1 : p != 0

from scipy.stats import spearmanr

iliski, pvalue = spearmanr(df["Ladder score"], df["Perceptions of corruption"])
print('Yolsuzluk Algısı Korelasyon Katsayısı = %.4f, P-Value = %.4f' % (iliski,pvalue))

iliski, pvalue = spearmanr(df["Ladder score"], df["Social support"])
print('Sosyal Destek Korelasyon Katsayısı = %.4f, P-Value = %.4f' % (iliski,pvalue))

iliski, pvalue = spearmanr(df["Ladder score"], df["Logged GDP per capita"])
print('GSYİH Korelasyon Katsayısı = %.4f, P-Value = %.4f' % (iliski,pvalue))

iliski, pvalue = spearmanr(df["Ladder score"], df["Freedom to make life choices"])
print('Yaşam Seçimleri Özgürlüğü Korelasyon Katsayısı = %.4f, P-Value = %.4f' % (iliski,pvalue))

iliski, pvalue = spearmanr(df["Ladder score"], df["Healthy life expectancy"])
print('Sağlıklı Yaşam Beklentisi Korelasyon Katsayısı = %.4f, P-Value = %.4f' % (iliski,pvalue))
```

* Tüm değişkenlerin p-value değeri 0.05'den küçük olduğu için H0 hipotezi reddedilir.
İstatistiksel olarak Korelasyon katsayıları anlamlıdır.

## Mutluluk Skorunu Etkileyen Faktörler;

* Mutluluk skorunu dikkate alındığında devletin sağladığı sosyal destek, sağlıklı yaşam beklentisi,
yaşam seçimlerini kişinin kendi karar verebilmesi özgürlüğü, kişi başına düşen milli gelir(GSYH) ile kuvvetli pozitif ilişki içinde bulunduğu gözükmektedir, **yani bu faktörlerde olumlu artışlar oldukça mutluluk skoru da aynı şekilde artmaktadır.**

* Yolsuzluk ile orta düzey negatif bir ilişki içinde, **yani yolsuzlukta olan artış mutluluk skorunu azaltmaktadır.**

* Cömertlik faktörünün ise mutluluk skoruna bir katkısı olmadığı görülmektedir.
